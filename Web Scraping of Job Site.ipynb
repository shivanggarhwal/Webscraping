```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://boston.craigslist.org/search/sof"
```

```python
response=requests.get(url)
```

```python
response
```

```python
data=response.text
data
```

```python
soup=BeautifulSoup(data,'html.parser')
```

```python
soup
```

```python
tags=soup.find_all('a')
```

```python
for tag in tags:
    print(tag.get('href'))
```

```python
titles=soup.find_all("a",{"class":"result-title"})
```

```python
for title in titles:
    print(title.text)
```

```python
addresses=soup.find_all("span",{"class":"result-hood"})
for address in addresses:
    print(address.text)
```

```python
ptags=soup.find_all("p",{"class":"result-info"})
for tag in ptags:
    title=tag.find("a",{"class":"result-title"}).text
    location_tag=tag.find("span",{"class":"result-hood"})
    location=location_tag.text[2:-1] if location_tag else "N/A"
    date=tag.find("time",{"class":"result-date"}).text
    link=tag.find("a",{"class":"result-title"}).get("href")
    job_response=requests.get(link)
    job_data=job_response.text
    job_soup=BeautifulSoup(job_data,"html.parser")
    job_decription=job_soup.find("section",{"class":"postingbody"})
    job_desription= job_decription.text if job_decription else "N/A"
    job_attributes_tag=job_soup.find("p",{"class":"attrgroup"})
    job_attributes=job_attributes_tag.text if job_attributes_tag else "N/A"
    print("title:",title,"\nlocation:",location,"\ndate:",date,"\nlink:",link,"\njob Description",job_decription,"\njob Attributes",
         job_attributes,"\n")
```

```python
job_no=0
while True:
    response=requests.get(url)
    data=response.text
    soup=BeautifulSoup(data,'html.parser')
    ptags=soup.find_all("p",{"class":"result-info"})
    for tag in ptags:
        title=tag.find("a",{"class":"result-title"}).text
        location_tag=tag.find("span",{"class":"result-hood"})
        location=location_tag.text[2:-1] if location_tag else "N/A"
        date=tag.find("time",{"class":"result-date"}).text
        link=tag.find("a",{"class":"result-title"}).get("href")
        job_response=requests.get(link)
        job_data=job_response.text
        job_soup=BeautifulSoup(job_data,"html.parser")
        job_decription=job_soup.find("section",{"class":"postingbody"})
        job_desription= job_decription.text if job_decription else "N/A"
        job_attributes_tag=job_soup.find("p",{"class":"attrgroup"})
        job_attributes=job_attributes_tag.text if job_attributes_tag else "N/A"
        job_no+=1
        print("title:",title,"\nlocation:",location,"\ndate:",date,"\nlink:",link,"\njob Description",job_decription,"\njob Attributes",
             job_attributes,"\n")
    url_tag=soup.find("a",{"title":"next page"})    
    if url_tag.get('href'):
        url="https://boston.craigslist.org/search/sof" +url_tag.get('href')
        print(url)
    else:
        break
print("total jobs", job_no)
```
